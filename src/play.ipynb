```python
def integrated_mass(x,c=1.0):
    m = log(1+x*c) - (x*c)/(1.0+(x*c))
    factor = log(1.0+c) - (c/(c+1.0))
    m = m/factor
    return m
```

```python
def integrated_vcirc(x,c=1.0):
    m = log(1+x*c) - (x*c)/(1.0+(x*c))
    factor = log(1.0+c) - (c/(c+1.0))
    m = m/factor
    m = sqrt(m/x)
    return m
```

```python
x = linspace(1E-2,1.0,200)
```

```python
plot(log10(x), log10(integrated_mass(x,c=1)))
plot(log10(x), log10(integrated_mass(x,c=4)))
plot(log10(x), log10(integrated_mass(x,c=10)))

#plot(x, integrated_mass(x,c=1))
```

```python
plot((x), (integrated_mass(x,c=1.0)))
plot((x), (integrated_mass(x,c=4)))
plot((x), (integrated_mass(x,c=20)))
plot((x), (integrated_mass(x,c=100)))
```

```python
plot((x), (integrated_vcirc(x,c=1.0)))
plot((x), (integrated_vcirc(x,c=2.0)))
plot((x), (integrated_vcirc(x,c=4.0)))
plot((x), (integrated_vcirc(x,c=10.0)))
```

```python
c_list = linspace(1.0,10.0,100)
x_vmax = empty((0))
for c in c_list:
    vcirc  = integrated_vcirc(x,c=c)
    v_max_id = argmax(vcirc)
    x_vmax = append(x_vmax, x[v_max_id])
scatter(c_list, x_vmax)
```
