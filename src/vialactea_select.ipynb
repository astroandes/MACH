```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("../data/halo_particles/ViaLactea/vl2subset_relposvel.txt")
```

```python
r = np.sqrt(data[:,0]**2 + data[:,1]**2 + data[:,2]**2)
n_part = len(r)
list_id = np.arange(n_part, dtype='int')
for i in range(5):
    list_id = np.random.permutation(list_id)

plt.plot(r[list_id])
```

```python
r.sort()
plt.plot(r)
```

```python
rho_crit = 1.48E-7 # Msun pc^-3
m_particle = 2.247E7 # Msun
```

```python
rho_enclosed = m_particle * (np.arange(n_part)+1)/(4.0/3.0 * np.pi * r**3)
rho_enclosed = rho_enclosed / rho_crit / 1E9
ii = rho_enclosed>200 # select below 200 times the critical density
r_vir = r[ii].max()
print r_vir
```

```python
plt.plot(np.log10(r[ii]), np.log10(rho_enclosed[ii]))
```

```python
plt.plot(np.log10(r), np.sqrt((np.arange(n_part)+1)/r))
```

```python
r = np.sqrt(data[:,0]**2 + data[:,1]**2 + data[:,2]**2)
ii = r < r_vir
x = data[ii,0]
y = data[ii,1]
z = data[ii,2]
new_data = x,y,z
new_data = np.transpose(new_data)
np.savetxt("../data/halo_particles/ViaLactea/vl2subset_virialcut.txt", new_data, fmt='%f')        
```
